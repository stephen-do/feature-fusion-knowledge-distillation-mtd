```python
import numpy as np
import matplotlib.pyplot as plt

# Load accuracy data from CSV files
acc5_loaded = np.loadtxt('acc5.csv', delimiter=',')
acc7_loaded = np.loadtxt('acc7.csv', delimiter=',')
acc9_loaded = np.loadtxt('acc9.csv', delimiter=',')
```
